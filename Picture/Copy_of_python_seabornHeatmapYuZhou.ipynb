```python
import pandas as pd
# import os
file = pd.read_csv("/content/drive/My Drive/20221029-Yu-HSV-HTDNA-transformed-aftermedianed.csv", index_col=0)
# file.head()
# print(file.columns)
# print(file.iloc[:, 4])
print(file.iloc[0:5,0:5])
# print(file.head())
# print(file2.iloc[0:2,0:2])
```

```python
import numpy as np
file2 = file.apply(np.log) #对数转换竟然同下面不一样，这个是自然对数
file3 = file2.iloc[0:54,]
print(file3.iloc[0:5, 0:5])
```

```python
df_log10 = np.log10(file) #对数转化https://cloud.tencent.com/developer/ask/sof/296062，这个是以10为底的对数
# print(df_log10.iloc[0:5, 0:5])
file3 = df_log10.iloc[0:54,]
print(file3.iloc[0:5, 0:5])
file3.describe()
```

```python
from seaborn.utils import axlabel
import matplotlib.pyplot as plt
sns.set_context({"figure.figsize":(20,20)})
# sns.heatmap(data=file3,square=True, vmin=-3,vmax=2, fmt='d', linewidths=.1, cmap='GnBu', annot=True)
ax = sns.heatmap(file3,  linewidths=.01, vmin=-4,vmax=4, cmap='Oranges', annot=True, annot_kws={'size':10, 'color':'black'}, cbar=True, cbar_kws={"shrink": 0.4})
plt.xlabel('Samples',fontsize=20, color='k') #x轴label的文本和字体大小
plt.ylabel('Metabolites',fontsize=20) #y轴label的文本和字体大小
plt.xlabel('x_label',fontsize=20, color='k') #x轴label的文本和字体大小
plt.ylabel('y_label',fontsize=20, color='k') #y轴label的文本和字体大小
plt.xticks(fontsize=10) #x轴刻度的字体大小（文本包含在pd_data中了）
plt.yticks(fontsize=10) #y轴刻度的字体大小（文本包含在pd_data中了）
plt.title('title',fontsize=20) #图片标题文本和字体大小

# colorbar该怎么设置？？
# cb = ax.figure.colorbar(ax.collections[0]) #显示colorbar
# cb.ax.tick_params(labelsize=10)#设置colorbar刻度字体大小
# cb.set_label('colorbar label') #设置colorbar的label



```
